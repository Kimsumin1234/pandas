### [실습1]

```python
# 실습을 위한 데이터 셋 로드 - 자동차

df = sns.load_dataset("mpg")
df
```

```python
# shape : 행 과 열 을 튜플형태로 리턴

print("행과열",df.shape)
print("행",df.shape[0])
```

### 3) sort_values()

```python
# mpg 컬럼을 기준으로 정렬하기
# 오름차순이 기본

df.sort_values('mpg')
```

```python
# mpg 컬럼을 기준으로 정렬하기 - 내림차순

df.sort_values('mpg',ascending=False)
```

### 4) rename

```python
# df.rename?
# Rename columns or index labels.
```

```python
# model_year 컬럼명 변경 : model_year => year

df.rename(columns = {'model_year':'year'})
```

### 6) reset_index() : index 를 컬럼으로 만들기(앞에서 사용함)

```python
# 인덱스가 기본으로 01234... 로 들어가있는 상태
df.head()

# 기본 인덱스 안쓰고 특정 컬럼을 인덱스로 지정
df.set_index("cylinders")

# 특정컬럼을 인덱스로 지정한걸 안쓰고 싶을때 사용
# cylinders 컬럼으로 지정한게 빠지면서 빠진 자리에 index 컬럼이 자동으로 생성
df.reset_index()
```

### 5) sort_index() : index 를 기준으로 정렬하기

```python
# sort_values(컬럼명) : 특정 컬럼을 기준으로 정렬시

df.sort_index(ascending=False)   # 위에서 만든 index 를 기준으로 정렬
```

### 7) drop(columns = ['','','']) : 괄호 안에 있는 컬럼 삭제 / drop(["",""],axis=1)

```python
# 중복 제거 : drop_duplicates() 는 duplicated() True/False 를 기반으로 돌아감

df.drop(['mpg','model_year'],axis=1)
```

```python
df.drop(columns = ['mpg','model_year'])  # df.drop(['mpg','model_year'],axis=1)
```

### [실습2]

```python
df = pd.DataFrame({"년도":[2019,2020,2021],"미국":[2.1,2.2,2.3], "한국":[0.4,0.5,0.45], "중국":[10,13,15]})
df
```

### 1) 인덱스 생성

```python
# 년도를 index로 생성하기

df = df.set_index("년도")
df
```

### 2) 인덱스 제거

```python
# 인덱스를 컬럼으로 변경하기

df = df.reset_index('년도')
df
```

### 3) 컬럼명 변경(중국==> 홍콩)

```python
df.rename(columns={'중국':'홍콩'},inplace=True)
df
```
