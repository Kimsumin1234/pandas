### 1) head / tail

````python
df = pd.DataFrame({
    "a" : [4,5,6,6,np.nan],
    "b" : [7,8,np.nan,9,9],
    "c" : [10,11,12,np.nan,12]
},
                  index=[1,2,3,4,5])
df
````

````python
# head() : Subset 의 대표적인 함수 (row 로 가져온다)

df.head()
````

````python
# tail() : Subset 의 대표적인 함수 (row 로 가져온다)

df.tail(2)
````

### 2) sample

````python
# sample : 랜덤으로 row 가져온다 (Subset)
# frac : 비율

df.sample(frac=0.5)
````

````python
# n의 개수는 전체의 행 수를 넘어서 지정하면 에러 발생
# 임의로 3개 가져오기 (n=3)

df.sample(n=3)
````

### 3) nlargest / nsmallest

````python
# a 컬럼에서 가장 큰거 3개, row 를 가져온다

df.nlargest(3, "a")
````

### 5) 불린 인덱싱

````python
friend_dict_list = [
    {"name":"John",
    "age" : 25,
    "job" : "student"},
    {"name":"Nate",
    "age" : 34,
    "job" : "teacher"},
    {"name":"Jenny",
    "age" : 30,
    "job" : "developer"}
]

friend_df = pd.DataFrame(friend_dict_list)
friend_df
````

````python
# 나이가 25세 초과인 사람

friend_df["age"] > 25 # pandas 는 for 구문을 사용안해도 그냥 해준다
friend_df.age > 25
````

````python
# [boolean 타입 조건을 넣는다]
# True 인 row 만 추출

friend_df[friend_df.age > 25]
````

````python
friend_df[friend_df.age > 30]
````

````python
# 나이 > 25 그리고 name = "Nate"
# pandas 는 and (X) ==> & 를 사용

# friend_df[(friend_df.age > 25) and (friend_df.name == "Nate")]
friend_df[(friend_df.age > 25) & (friend_df.name == "Nate")]
````

````python
friend_df[(friend_df.age > 25) & (friend_df.name == "Jenny")]
````

### 6) query()

````python
friend_df.query("age > 25")
````

````python
# query 는 and 사용가능

friend_df.query("age > 25" and "name == 'Nate'")
````

#### [실습] drop_duplicates()

````python
student_list = pd.DataFrame(
    {
        "name" : ["John","Nate","Edward","Zara","John"],
        "major" : ["Computer Science","Computer Science","Computer Science","Psychology","Computer Science"],
        "gender" : ["male","male","male","female","male"]
    }
)
student_list
````

````python
student_list.drop_duplicates()
````

````python
# drop_duplicates() : duplicated() 가 True 인 row 를 삭제
# duplicated() 로 중복이 있는지 먼저 확인가능하다

student_list.duplicated()
````

````python
student_list = pd.DataFrame(
    {
        "name" : ["John","Nate","Edward","Zara","Wendy","Nate","John"],
        "major" : ["Computer Science","Computer Science","Computer Science","Psychology","Economics","None","Economics"],
        "gender" : ["male","male","male","female","female","male","male"]
    }
)
student_list
````

````python
# 기본 : row 전체가 완벽히 중복인걸 판별

student_list.duplicated()
````

````python
# 조건부여 가능 : 이름이 동일할때 중복이라고 인식 해보기

student_list.duplicated("name")
````

````python
# keep : 중복 데이터 중에서 어느 행을 살릴 것인가? (기본값은 first)

student_list.drop_duplicates("name", keep="last")
````
