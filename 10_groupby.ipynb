```python
import pandas as pd
import seaborn as sns
import numpy as np
```

### [실습1]

```python
df = sns.load_dataset("mpg")
df.head()
```

```python
# 개수 세기

df.groupby("origin") # DataFrameGroupBy object 타입
df.groupby("origin").size()
```

```python
# value_counts(컬럼명)

df.value_counts('origin')
```

```python
# origin 별로 평균값을 보고 싶다면?

# numeric_only=True : 문자가 포함된 경우 숫자만 평균 구해야해서 이걸 같이 사용한다
df.groupby("origin").mean(numeric_only=True)
```

```python
# cylinders 값만 보고 싶다면?

df.groupby("origin")['cylinders'].mean()
```

```python
# origin 별 cylinders 중간값 보기

df.groupby("origin")['cylinders'].median()
```

```python
# 그룹을 여러개 설정

df.groupby(['model_year',"origin"])['cylinders'].mean()
```

```python
df = pd.DataFrame(
    {
        'name':['John','Nate','Abraham','Brian','Janny','Yuna','Jeniffer','Edward','Zara','Wendy','Sera'],
        'major':['Computer Science','Computer Science','Physics','Psychology','Economics','Economics','Computer Science','Computer Science','Psychology','Economics','Psychology'],
        'gender':['male','male','male','male','female','female','female','male','female','female','female'],
    }
)

df
```

### transform()
- transform(집계함수) : 연산의 결과를 현재 데이터프레임의 인덱스에 맞춰서 돌려준다

```python
# df_car.groupby('Branch')['Ext Price'].sum()
df_car.groupby('Branch')['Ext Price'].transform('sum')
```

```python
# merge() 를 안하고 transform() 을 사용

df_car['Br_Total'] = df_car.groupby('Branch')['Ext Price'].transform('sum')
df_car
```

```python
df = pd.DataFrame(
    {
        'School':['Yeonhi','Yeonhi','Sungsan','Sungsan','Sungsan'],
        'Name':['Haena','Gisu','Una','Naeun','Ziho'],
        'Math_s':[92,70,88,92,70],
    }
)

df
```

```python
# school을 그룹으로 하여 수학 성적의 평균 구하기

# df_m = df.groupby('School')['Math_s'].mean().reset_index().rename(columns={'Math_s':'Avg_s'})
df_m = df.groupby('School')['Math_s'].mean().rename('Avg_s').reset_index()
df_m
```

```python
df = df.merge(df_m)
df
```

- transform() 을 사용시

```python
df['Avg_s'] = df.groupby('School')['Math_s'].transform('mean')
df
```

- apply() : apply(적용할함수)

```python
# 점수가 90 이상이면 A, 80 이상이면 B, 70 이상이면 C, 나머지 F
def grade(num):
  if num >= 90:
    result = 'A'
  elif num >= 80:
    result = 'B'
  elif num >= 70:
    result = 'C'
  else:
    result = 'F'

  return result
```

```python
# 함수동작 확인

grade(80)
```

```python
df
```

```python
df['grade'] = df['Math_s'].apply(grade) # map(grade), applymap(X)
df
```

```python
df['grade'] = df['Math_s'].apply(grade)
df_m = df
df_m
```

```python
df_m.sort_values('Math_s',ascending=False)
```

```python
def sort_math(df,column='Math_s',n=3):
  return df.sort_values(column,ascending=False)[:n]
```

```python
df_m.groupby('School').apply(sort_math,n=2)
```
