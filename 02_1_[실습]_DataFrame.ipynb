### [실습1]

```python
import pandas as pd
```

```python
dict = {"홍길동":[88,75,95],"강감찬":[82,62,85],"김종민":[75,69,70],"임창정":[45,56,80]}
table = pd.DataFrame(dict)
table
```

```python
table.index = ["java","python","basic"]
table
```

```python
# 홍길동 학생의 과목별 점수를 출력하시오
table["홍길동"]
```

```python
# python 과목의 점수를 출력하시오.
# 행 은 loc 사용

table.loc["python"]
```

```python
# 각 학생의 sql 점수를 추가하시오 [70,85,86,92]
table.loc["sql"] = [70,85,86,92]
table
```

```python
# basic 점수를 삭제하시오
table.drop("basic",inplace=True)
table
```

### [실습2]

```python
df = pd.DataFrame({"미국":[2.1,2.2,2.3], "한국":[0.40,0.50,0.45], "중국":[10,13,15]})
df
```

```python
# 인덱스와 함께 생성하기
df = pd.DataFrame({"미국":[2.1,2.2,2.3], "한국":[0.40,0.50,0.45], "중국":[10,13,15]}, index=[2019,2020,2021])
df
```

```python
#  인덱스를 출력하시오
df.index
```

```python
# 컬럼명을 출력하시오.
df.columns
```

```python
# 전체 값들만 출력하기
df.values
```

```python
# 인덱스로 특정 컬럼 선택하기
df["미국"]
```

```python
df.미국 # 한글일 경우는 . 접근방식이 안될수도 있다 그래서 ["미국"] 사용을 추천
```

```python
# 일본 컬럼 추가하기
df["일본"] = [3,4,5]
df
```
