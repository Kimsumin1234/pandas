#### 주어진 코드를 이용하여 퀴즈를 풀어보세요

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "영화" : ['명량','극한직업','신과함께 -죄와 벌','국제시장', '베테랑', '괴물', '도둑들', '7번방의 선물', '암살', '광해, 왕이 된 남자'],
    "개봉 연도" : [2014, 2019, 2017, 2014, 2015, 2006, 2012, 2013, 2015, 2012],
    "관객 수" : [1761, 1626, 1441, 1425, 1341, 1301, 1298, 1281, 1270, 1231],  # 단위 : 만명
    "평점" : [8.88, 9.20, 8.73, 9.16, 9.24, 8.62, 7.64, 8.83, 9.10, 9.25]
}
```

```python
# 데이터 프레임 생성
df = pd.DataFrame(data)
df
```

```python
df['영화']
```

#### 4) 주어진 계산식을 참고하여 '추천점수' 컬럼 추가

### [퀴즈5]

#### 1) covid/04-01-2020.csv 파일 dataframe 생성 (데이터프레임 변수 이름은 covid_df 로  지정)

```python
covid_df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/data/covid/04-01-2020.csv')
covid_df.head(3)
```

#### 2) covid_df 에서 중복된 Country_Region 각각에 대해 Confirmed 를 모두 더한 값을 컬럼으로 갖는 데이터프레임을 피봇테이블로 만드세요

```python
# 각나라별 확진자 수
# pivot() : 함수 적용 안됨
# pivot_table() : 함수 적용 가능

pd.pivot_table(covid_df, index='Country_Region', values='Confirmed', aggfunc='sum')
```

#### 3) covid_df 에서 중복된 Country_Region 각각에 대해 Confirmed 를 모두 더한 값을 컬럼으로 갖는 데이터프레임을 피봇테이블로  만들고, 총 값을 'Total' 을 인덱스 값으로 하는 record 로 추가하세요
- 결측치는 0으로 처리

```python
# row Total 추가하기 : margins=True, margins_name="Total"
# 결측치 0 : fill_value=0 (dropna= 같은거도 있음)
pd.pivot_table(covid_df, index='Country_Region', values='Confirmed', aggfunc='sum', margins=True, margins_name="Total", fill_value=0)
```
