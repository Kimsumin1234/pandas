#### NaN(Not a Number) - 표현 불가능한 데이터(비어 있는 값)

- NaN : missing value 를 표현하는 기본 형태
- 기본적으로 float 형식으로 처리됨

#### NA(Not Available) : 결측값
#### None : 값의 부재(값이 존재하지 않거나, 없음, 정의되지 않음)

```python
import pandas as pd
import numpy as np
```

### [실습 1]

#### 1) missing data 가 포함된 데이터 프레임 생성

```python
df = pd.DataFrame(
    {
        "name":["Alfred","Batman","Catwoman"],
        "toy":[np.nan,"Batmobile","Bullwhip"],
        "born":[None,pd.Timestamp("19400425"),pd.NA]
    }
)

df
```

#### 2) 데이터 타입 확인

```python
df.dtypes
```

#### 3) missing data 처리

**dropna : missing values 제거**

```python
# df.dropna?
```

df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

```python
# how=any 행 or 열에 하나라도 널 값이 있으면 없애기
# axis = 0(행), 1(열)

df.dropna(axis=1)
```

```python
# how=any 행 or 열에 하나라도 널 값이 있으면 없애기
# axis = 0(행), 1(열) , 기본값이 0

df.dropna()
```

```python
# how='all' 전체 행에 결측치가 모두 있어야 drop
# 기본값 이 how='any'

df.dropna(how='all')
```

**fillna : missing values 를 임의의 값으로 채우기**

```python
# 널값 채우기

# df.fillna?
```

df.fillna(
    value=None,
    method=None,
    axis=None,
    inplace=False,
    limit=None,
    downcast=None,
)

```python
# 결측치를 무엇으로 채울 것인가?
# scalar : 단일 값

df.fillna(0)
```

```python
# 특정 값으로 채우기 - 값이 NaN, NaT, NA, None인 경우에만 채워짐

values = {"name":"noname", "toy":"Bat", "born":pd.Timestamp("19000101")}
```

```python
df.fillna(value=values)
```

### [실습 2]

```python
student_list = {
    "name":
    ["John", "Nate", "Yuna", "Abraham", "Brian", "Janny", "Nate", "John"],
    "job": [
        "teacher", "teacher", "teacher", "student", "student", "student",
        "teacher", "student"
    ],
    "age": [40, 35, 37, 10, 12, 11, None, None]
}

df = pd.DataFrame(student_list)
df
```

```python
# 전체 행의 개수와 컬럼 개수 확인하기

df.shape
```

```python
# 전체적인 구조 확인하기
# info()
# name    8 non-null => name 에서 8개만 not Null 임을 알려준다
# age     6 non-null => null 이 있으면 알아서 개수에서 빠진다

df.info()
```

```python
df.isnull()
```

```python
# 없는 값을 0 으로 채우기

df.age = df.age.fillna(0)
df
```

```python
# job 을 그룹으로 잡고 age 에 대해서 평균 구하기

df.groupby("job")["age"].transform("median")
```

```python
# 평균 값으로 채우기

df["age"].fillna(df.groupby("job")["age"].transform("median"), inplace=True)
df
```

### 학생 실습

```python
df = pd.DataFrame(
    [
        [np.nan, 2, np.nan, 0],
        [3, 4, np.nan, 1],
        [np.nan, np.nan, np.nan, 5]
    ], columns=list("ABCD")
)

df
```

```python
# 결측치 확인
df.isnull()
```

```python
# 모든 결측치 0 으로 채우기
df.fillna(0)
```

```python
# 모든 결측치 특정 값으로 채우기
# A:0 B:1 C:2 D:3
df.fillna({"A":0, "B":1, "C":2, "D":3})
```

```python
# 모든 결측치 D열의 중간값(median)으로 채우기

df.fillna(df["D"].median())
```

```python
# 모든 결측치 D열의 최대값(max)으로 채우기

df.fillna(df["D"].max())
```

### [개념] NaN

- NaN 을 나타내는 방법으로 np.nan, pd.NaT, pd.NA, None 등이 있음
- 단, np.nan 은 개별 값으로 비교 시 비교가 안됨
- None 는 비교가 됨

```python
# None 만 비교가 가능

print("np.nan != np.nan : ", np.nan != np.nan)
print("np.nan == np.nan : ", np.nan == np.nan)
print("pd.NaT == pd.NaT : ", pd.NaT == pd.NaT)
print("pd.NA == pd.NA : ", pd.NA == pd.NA)
print("None == None : ", None == None)
```

- 표현할 수 없는 값들을 서로 비교를 하기 원한다면 None 로 처리하는 것이 편함
