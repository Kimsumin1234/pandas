### 삭제
* 수정과 삭제는 해당 dataframe에 적용을 하지 않으면 반영 안됨
* 직접 반영을 위해 inplace=True 사용/ df에 직접 변경을 원하지 않으면 새로운 변수에 담아서 사용하기

```python
# drop()
# 기본형태는 결과값을 새로 담지 않으면 원본파일은 변경되지 않음

friend_df.drop({0,2})
```

```python
friend_df
```

```python
# inplace=True 를 사용하면 원본파일을 바로 변경가능

friend_df.drop({0,2},inplace=True)
```

```python
friend_df
```

```python
friend_df = pd.DataFrame(friend_dict_list)
friend_df
```

```python
# 컬럼명을 사용해서 삭제
# drop("제거할데이터",axis=0) : axis 기본값 0 (=행), 삭제방향이 행이 기준임(기본)

friend_df.drop("age")
```

```python
# 삭제방향 변경하면 컬럼명으로 삭제 가능

friend_df.drop("age",axis=1)
```

#### 3) 컬럼 추가

```python
# append 개념, 뒤에 붙는다

friend_df["주소"] = ["서울","경기","부산"]
friend_df
```

```python
# loc[행레이블] = [데이터]

friend_df.loc["p4"] = ["Tom","010-7891-4578",28,"tester","대전"]
friend_df
```
